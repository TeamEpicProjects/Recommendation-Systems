```python
import pandas as pd
import gzip
import json

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l.decode('utf-8'))

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('./Luxury_Beauty.json.gz')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = data['description'].values.astype("U")

vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
wss=[]
for i in range(1,20):
    model = KMeans(n_clusters=i, init='k-means++', max_iter=100, n_init=1)
    model.fit(features)
    wss.append(model.inertia_)
```

### Similarity Check

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

extras=['benefits<br />','','a','this','that','these','those','to','of','at','with','for','also','is']

stop.append(extras)
```

```python
from sklearn.metrics.pairwise import linear_kernel

def similarity(result, title):
    desc_vector = tfv.fit_transform(result['description'].apply(lambda x:x.lower()).apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)])))
    similarity_matrix = linear_kernel(desc_vector, desc_vector)
    
    mapping = pd.Series(result.index, index=result['title'])
    product_index = mapping[title]
    
    similarity_score = list(enumerate(similarity_matrix[product_index]))
    
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    
    list_similarity = []
    for i in range(len(similarity_score)):
        list_similarity.append(similarity_score[i][1])
    
    result['similarity'] = list_similarity
    
    return result
```

### Getting recommendations

```python
def get_rec_without_clustering(title):
    
    i=indices[title]
    sig_scores=list(enumerate(sig[i]))
    
    scores=sorted(sig_scores, key=lambda x:x[1], reverse=True)
    
    scores=scores[1:11]
    
    product_indices=[i[0] for i in scores]
    
    return data.iloc[product_indices]
```

```python
def get_rec(title,sig=sig):
    i=indices[title]
    x=data.iloc[i]['label']
    t=[x]
    idx=list(data[data['label'].isin(t)].index)
    sig_temp=list(enumerate(sig[i]))

    sig_scores=itemgetter(*idx)(sig_temp)
    
    scores=sorted(sig_scores, key=lambda x:x[0], reverse=True)
    
    product_indices=[i[0] for i in scores]
    
    result = data.iloc[product_indices]
    result=result.reset_index(drop=True)

    output = similarity(result, title)
    output.drop(['label','similarity'], axis=1, inplace=True)
    
    return output[:20]
```

```python
import pandas as pd
import gzip
import json

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l.decode('utf-8'))

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('./Luxury_Beauty.json.gz')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = data['description'].values.astype("U")

vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
wss=[]
for i in range(1,20):
    model = KMeans(n_clusters=i, init='k-means++', max_iter=100, n_init=1)
    model.fit(features)
    wss.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')

plt.plot(range(1,20), wss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances')
plt.show()
```

```python
def get_rec_without_clustering(title):
    
    i=indices[title]
    sig_scores=list(enumerate(sig[i]))
    
    scores=sorted(sig_scores, key=lambda x:x[1], reverse=True)
    
    scores=scores[1:11]
    
    product_indices=[i[0] for i in scores]
    
    return data.iloc[product_indices]
```

```python
def get_rec(title,sig=sig):
    
    i=indices[title]
    t=data.iloc[i]['label']
    
    idx=list(data[data['label']==t].index)
    
    sig_temp=itemgetter(*idx)(sig)
    sig_scores=list(enumerate(sig_temp[i]))
    
    scores=sorted(sig_scores, key=lambda x:x[1], reverse=True)
    
    scores=scores[1:21]
    
    product_indices=[i[0] for i in scores]
    
    return data.iloc[product_indices]
```
